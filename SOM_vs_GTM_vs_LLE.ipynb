# Self Organizing Maps and Local Linear Embedding on a Swiss Roll (data from Tenenbaum et al)

## Imports

```python
%matplotlib inline
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
Axes3D
from sklearn import manifold, datasets
import pandas
import numpy  as np
import pandas as pd
import scipy.io as io
```

## Load swiss roll data

```python
swiss_roll= io.loadmat("/Users/Vishakh/Desktop/Som_lle_gtm project/Datasets/swiss_roll.mat")
X= np.array(swiss_roll['X_data'])
Yfull=np.array(swiss_roll['Y_data'])
X=X.transpose()
Yfull=Yfull.transpose()
Y=Yfull[:,0]
xpt,ypt,zpt=X[:, 0], X[:, 1], X[:, 2]
Yfull.shape
```

```python
plt.scatter(Yfull[:,0],Yfull[:,1],c=Yfull[:,0],cmap=plt.cm.Spectral)
```

```python
app_loc=Yfull[:,0]
grouping = ['']*20000
for i in range(len(app_loc)):
    
    if  app_loc[i]>=0 and app_loc[i] < 20 :
        grouping[i]='1'
    if  app_loc[i]>=20 and app_loc[i]<40:
        grouping[i] = '2'
    if  app_loc[i]>=40 and app_loc[i]<60 :
        grouping[i] = '3'
    if  app_loc[i]>=60 and app_loc[i]<80 :
        grouping[i] = '4' 
    if app_loc[i] >=80 and app_loc[i]<100 :
        grouping[i] = '5' 
  
  
    
    
        
grouping= np.array(grouping)
```

```python
one = app_loc[app_loc>0]
two = one[one<10]
len(two)
```

## Plot the swiss roll and then the dimensionally reduced points via LLE

```python


X_r, err = manifold.locally_linear_embedding(X, n_neighbors=12,
                                             n_components=2)


fig = plt.figure(figsize=(20,10))
try:
    # compatibility matplotlib < 1.0
    ax = fig.add_subplot(211, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=Y, cmap=plt.cm.Spectral)
except:
    ax = fig.add_subplot(211)
    ax.scatter(X[:, 0], X[:, 2], c=Y, cmap=plt.cm.Spectral)

ax.set_title("Original data")
ax = fig.add_subplot(212)
ax.scatter(X_r[:, 0], X_r[:, 1], c=Y, cmap=plt.cm.Spectral)
plt.axis('tight')
plt.xticks([]), plt.yticks([])
plt.title('Projected data')
plt.show()
```
