```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
Axes3D
%matplotlib inline
from sklearn import manifold, datasets
import numpy as np
```

# LSUN

```python
path = "/Users/Vishakh/Desktop/Som_lle_gtm project/FCPS/01FCPSdata/Lsun.lrn"
path2 = "/Users/Vishakh/Desktop/Som_lle_gtm project/FCPS/01FCPSdata/Lsun.cls"
data = np.loadtxt(path,skiprows=4)
data = data[:, [1, 2]]
colors = np.loadtxt(path2,skiprows=4)
colors= colors[:,1]
colors=np.array(colors)
colors=colors.astype(int)
colormap = np.array(['x', 'g', 'b','r'])
fig = plt.figure(figsize=(20,10))
#ax = fig.add_subplot(211, projection='3d')
#ax.scatter(data[:, 0], data[:, 1], data[:, 2],c=colormap[colors])
plt.scatter(data[:, 0], data[:, 1],c=colormap[colors])
```

```python
X_r, err = manifold.locally_linear_embedding(data, n_neighbors=20,
                                             n_components=2)
```

```python
plt.scatter(X_r[:, 0], X_r[:, 1], c=colormap[colors], cmap=plt.cm.Spectral)
```

```python
err
```

```python
%load_ext rmagic
%Rpush data
%Rpush colors
```

```python
%%R
library('popsom')
m<-map.build(data,labels=colors,xdim=25,ydim=20,train=200000,alpha=0.8)
map.starburst(m)
```

```python
%%R
print(map.convergence(m))
```
