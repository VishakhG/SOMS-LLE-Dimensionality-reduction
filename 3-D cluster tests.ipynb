# Hepta 

```python
path = "/Users/Vishakh/Desktop/Som_lle_gtm project/FCPS/01FCPSdata/Hepta.lrn"
path2 = "/Users/Vishakh/Desktop/Som_lle_gtm project/FCPS/01FCPSdata/Hepta.cls"
data = np.loadtxt(path,skiprows=4)
data = data[:, [1,2,3]]
colors = np.loadtxt(path2,skiprows=4)
colors= colors[:,1]
colors=np.array(colors)

set1=data[colors==1]
set2=data[colors==2]
set3=data[colors==3]
set4=data[colors==4]
set5=data[colors==5]
set6=data[colors==6]
set7=data[colors==7]

colormap = np.array([ 'b','g','r','c','m','y','k','w'])
colors=colors.astype(int)

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(211, projection='3d')
#ax.scatter(data[:, 0], data[:, 1], data[:, 2],norm=norm,c=colors,cmap = plt.jet)
ax.plot(set1[:,0],set1[:,1],set1[:,2],'o',color=colormap[1])
ax.plot(set2[:,0],set2[:,1],set2[:,2],'o',color=colormap[2])
ax.plot(set3[:,0],set3[:,1],set3[:,2],'o',color=colormap[3])
ax.plot(set4[:,0],set4[:,1],set4[:,2],'o',color=colormap[4])
ax.plot(set5[:,0],set5[:,1],set5[:,2],'o',color=colormap[5])
ax.plot(set6[:,0],set6[:,1],set6[:,2],'o',color=colormap[6])
ax.plot(set7[:,0],set7[:,1],set7[:,2],'o',color=colormap[7])
```

```python
X_r, err = manifold.locally_linear_embedding(data, n_neighbors=65
                                             ,
                                             n_components=2)

plt.scatter(X_r[:, 0], X_r[:, 1],c=colormap[colors])
```
